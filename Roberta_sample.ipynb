```python
#importing necessary libraries
import numpy as np
import pandas as pd
import transformers
import torch
from transformers import AutoTokenizer, AutoModelForQuestionAnswering, pipeline
```

```python
# Specifying huggingface's Roberta model
model_name = 'deepset/roberta-base-squad2'
```

```python

# Dicts requires an input question. This question has to be in the scope of the context
dicts = [ {'Questions': input('Enter Question:'),
          'Context': 'Energy is the ability to do work. There are many different forms of energy, including kinetic energy, potential energy, thermal energy, radiant energy, and chemical energy. Energy can be transferred from one form to another, but it cannot be created or destroyed. Kinetic energy is the energy of motion. The faster an object is moving, the more kinetic energy it has.Potential energy is the energy stored in an object due to its position or configuration. A stretched spring has potential energy, as does a book held above the ground. Thermal energy is the energy of motion of the atoms and molecules in an object. The hotter an object is, the more thermal energy it has. Radiant energy is the energy carried by electromagnetic waves, such as light. Chemical energy is the energy stored in the chemical bonds of a substance.'}]
```

```python

# Initializing model and tokenizer
model = AutoModelForQuestionAnswering.from_pretrained(model_name)
tokenizer = AutoTokenizer.from_pretrained(model_name)
```

```python

# Tokenize and encode the inputs
inputs = tokenizer(dicts[0]['Questions'], dicts[0]['Context'], return_tensors='pt')
# The tokenizer method tokenizes and encodes the question and context.
# The 'dicts[0]['Questions']' represents the question, and 'dicts[0]['Context']' represents the context.
# 'return_tensors='pt'' specifies that the tokenizer should return PyTorch tensors.

# Perform question answering
outputs = model(**inputs)
# The model method takes the inputs and performs question answering.
# The '**' is used to unpack the 'inputs' dictionary as keyword arguments.
```

```python
outputs
```

```python

# Find the index of the highest start logit
ansr_strt_idx = torch.argmax(outputs.start_logits)

# Find the index of the highest end logit
ansr_end_idx = torch.argmax(outputs.end_logits)
```

```python

# Retrieve the answer span tokens from the encoded inputs
ansr_tokens = inputs.input_ids[0, ansr_strt_idx:ansr_end_idx+1]

# Decode the answer span tokens into a human-readable string
answer = tokenizer.decode(ansr_tokens)
```
